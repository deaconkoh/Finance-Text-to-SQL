```python
from datasets import load_dataset

from data_utils import (
    load_finch_schemas,
    get_finch_db_path,
    inspect_sqlite_tables,
    get_matching_schema,
    format_schema_for_prompt,
    get_full_schema_cached,
)
```

```python


dataset = load_dataset("domyn/FINCH")
sample = dataset["train"][0]

print("=== FINCH Sample ===")
print(f"Question: {sample['question']}")
print(f"SQL: {sample['SQL']}")
print(f"Database: {sample['db_id']}")
print(f"Database source: {sample['db_name']}")
print(f"Partition: {sample['partition']}")
print(f"Difficulty: {sample['difficulty']}")

print("\n=== Load Schema YAML ===")
schemas = load_finch_schemas()
print(f"Schema object type: {type(schemas)}")
print(f"Number of schema items: {len(schemas)}")

print("\n=== Download Matching SQLite DB ===")
db_path = get_finch_db_path(sample)
print(f"DB path: {db_path}")

print("\n=== SQLite Tables ===")
tables = inspect_sqlite_tables(db_path)
print(f"Query Results: {tables}")

print("\n=== Matching Schema ===")
matching_schema = get_matching_schema(sample, schemas)
print(type(matching_schema))
print(matching_schema.keys())

print("\n=== Formatted Schema Text ===")
schema_text = format_schema_for_prompt(matching_schema)
print(schema_text)

print("\n=== Cached Schema Text Check ===")
cached_schema_text = get_full_schema_cached(sample, schemas)
print(cached_schema_text[:1000])
```
